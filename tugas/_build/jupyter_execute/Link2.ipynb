```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import sys

# Fungsi untuk menampilkan progress bar
def print_progress(current, total):
    percent = (current / total) * 100
    bar_length = 20
    filled_length = int(bar_length * current // total)
    bar = '‚ñà' * filled_length + '-' * (bar_length - filled_length)
    sys.stdout.write(f'\r[{bar}] {percent:.2f}%')
    sys.stdout.flush()
    if current == total:
        sys.stdout.write('\n\n')

def scrape_links():
    start_time = time.time()  # mulai hitung waktu
    data = {
        "no": [],
        "page": [],
        "link_keluar": []
    }
    
    no = 1  # nomor urut
    # Misalnya kita mulai dari Link A
    link_A = "https://pta.trunojoyo.ac.id/"
    visited_links = set()  # untuk memastikan kita tidak mengunjungi link yang sama berulang kali

    # Loop untuk link A ke B, B ke C, C ke A dan C ke B
    for i in range(1, 42):  # misalnya ada 42 halaman kategori
        total_pages = 3  # misalnya ada 3 halaman per kategori
        for j in range(1, total_pages + 1):  # loop untuk setiap halaman kategori
            url_A = f"{link_A}/c_search/byprod/{i}/{j}"
            r = requests.get(url_A)
            soup = BeautifulSoup(r.content, "html.parser")
            jurnals = soup.select('li[data-cat="#luxury"]')

            isii = soup.select_one('div#begin')
            if not isii:
                continue
            prodi_full = isii.select_one('h2').text.strip()
            prodi = prodi_full.replace("Journal Jurusan ", "")

            for jurnal in jurnals:
                link = jurnal.select_one('a.gray.button')['href']

                # Menambahkan data
                data["no"].append(no)
                data["page"].append(url_A)          # link page
                data["link_keluar"].append(link)  # link detail
                no += 1

            # Update progress bar
            print_progress(j, total_pages)

    # Simpan data dalam DataFrame dan export ke CSV
    df = pd.DataFrame(data)
    df.to_csv("scraped_links.csv", index=False)

    end_time = time.time()
    elapsed = int(end_time - start_time)
    jam, sisa = divmod(elapsed, 3600)
    menit, detik = divmod(sisa, 60)

    # Summary
    print("\n‚úÖ Link berhasil dikumpulkan!")
    print(f"üìä Total entri: {len(df)}")
    print(f"‚è±Ô∏è Waktu eksekusi: {jam} jam {menit} menit {detik} detik")

    return df

# Memanggil fungsi scraping
scrape_links()
```
